### Replacing column name (Measure:volume >> subject), modifying his values and setting 'subject' as index

#### ALFA_PLUS

```python
#renaming values of Measure:volume column

names = []
scans = []
for path, subdirs, files in os.walk(r'/home/jpozoc/barracuda/jpozoc/Archivos_ALFAPLUS_security_copy'):
    sorted_files = sorted(files)
    for filename in sorted_files:
        names.append(int(filename[0:5]))
        scans.append(int(filename.split("_")[2]))
for i in range(0,statsFastS.shape[0]):
    statsFastS["Measure:volume"].replace({i: names[i]}, inplace=True)

#statsFastS['Scan'] = scans    
#rename column    
statsFastS.rename(columns={'Measure:volume': 'subject'}, inplace=True)

#set subject as index
statsFastS.set_index('subject', inplace= True)

statsFastS.head()
```

#### ALFA_OPCIONAL

```python
#renaming values of Measure:volume column

names = []
#scans = []
for path, subdirs, files in os.walk(r'/home/jpozoc/barracuda/jpozoc/ALFA_OPCIONAL_security'):
    sorted_files = sorted(files)
    for filename in sorted_files:
        names.append(int(filename[0:5]))
        #scans.append(int(filename.split("_")[2]))
for i in range(0,statsFastS_AO.shape[0]):
    statsFastS_AO["Measure:volume"].replace({i: names[i]}, inplace=True)

#statsFastS['Scan'] = scans    
#rename column    
statsFastS_AO.rename(columns={'Measure:volume': 'subject'}, inplace=True)

#set subject as index
statsFastS_AO.set_index('subject', inplace= True)

statsFastS_AO.head()
```

### Checking common columns in both dataframes (if they are not common, they will be removed)

#### ALFA_PLUS

```python
#if a column is not in the other dataframe, it is removed

column_notFastS = []
for column in statsFastS.columns:
    if column == 'Scan':
        continue
    elif column not in statsFreeS.columns:
        print(column)
        column_notFastS.append(column)
    else:
        continue
statsFastS.drop(column_notFastS,axis=1, inplace=True)

column_notFreeS = []
for column in statsFreeS.columns:
    if column not in statsFastS.columns:
        print(column)
        column_notFreeS.append(column)
    else:
        continue
statsFreeS.drop(column_notFreeS,axis=1, inplace=True)
```

#### ALFA_OPCIONAL

```python
#if a column is not in the other dataframe, it is removed

column_notFastS = []
for column in statsFastS_AO.columns:
    if column == 'Scan':
        continue
    elif column not in statsFreeS_AO.columns:
        print(column)
        column_notFastS.append(column)
    else:
        continue
statsFastS_AO.drop(column_notFastS,axis=1, inplace=True)

column_notFreeS = []
for column in statsFreeS_AO.columns:
    if column not in statsFastS_AO.columns:
        print(column)
        column_notFreeS.append(column)
    else:
        continue
statsFreeS_AO.drop(column_notFreeS,axis=1, inplace=True)
```

### Removing those rows that are duplicated. We print the index location and then we remove them

#### ALFA_PLUS

```python
lista = statsFastS.index.duplicated().tolist()
for i in [i for i,x in enumerate(lista) if x == 1]:
    print(i)
    
statsFastS = statsFastS[~statsFastS.index.duplicated(keep='first')] #removing duplicated subjects
```
