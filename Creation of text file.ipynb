### Creación de función para crear una carpeta por cada archivo, y guardar este archivo dentro de la carpeta creada

```python
import glob
from pathlib import Path

def freesurfer_prep(directory:str, file:str, final_name:str=None)->list:
    Path(directory).mkdir(parents=True, exist_ok=True)
    os.rename(file, os.path.join(directory, final_name))

[freesurfer_prep(directory=f.split('.')[0], file=f, final_name='orig.nii.gz') for f in glob.glob('/media/hdd1/jpozoc/ALFA_OPCIONAL/*')]
```
